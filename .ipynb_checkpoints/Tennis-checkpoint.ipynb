```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_excel('2010_mens.xlsx')
```

```python
df.shape
```

```python
df.tourney_date = pd.to_datetime(df.tourney_date, yearfirst=True, format='%Y%m%d')
df = df.sort_values(by='tourney_date').reset_index(drop=True)
```

```python
df.shape
```

```python
df
```

```python
data = np.ones((len(df), 1))*1500

df['winner_elo'] = data
df['loser_elo'] = data
```

```python
sub_df = df[['winner_name', 'loser_name', 'winner_elo', 'loser_elo']]
sub_df
```

```python
sub_df.isnull().sum()
```

```python
def probability(elo_1, elo_2):
    
    return 1/(1 + math.pow(10, (elo_1 - elo_2)/400))
```

```python
def elo_rating(rating_player, rating_opponent, K):
    
    prob_a = probability(rating_opponent, rating_player)
    prob_b = probability(rating_player, rating_opponent)
    
    rating_player = rating_player + K * (1 - prob_a)
    rating_opponent = rating_opponent + K * (0 - prob_b)
    
    return rating_player, rating_opponent
```

```python
sub_df.shape
```

```python
elo_winner, elo_loser = elo_rating(sub_df.winner_elo[670], sub_df.loser_elo[670], 30)
elo_winner
```

```python
df.at[3, 'winner_elo'] = 1700
df.at[3, 'winner_elo']
```

```python
len(df[df['winner_name']=='Novak Djokovic'].iloc[43:, :].winner_elo.values.tolist())
```

```python
for i in range(df.shape[0]):
    elo_winner, elo_loser = elo_rating(df.at[i, 'winner_elo'], df.at[i, 'loser_elo'], 30)
    
    df.at[i, 'winner_elo'] = elo_winner
    df.at[i, 'loser_elo'] = elo_loser
    
    winner_name = df.winner_name[i]
    loser_name = df.loser_name[i]
    
    sub_ww = df[df['winner_name']==winner_name].winner_elo.iloc[i:]
    sub_ll = df[df['loser_name']==loser_name].loser_elo.iloc[i:]
    sub_lw = df[df['loser_name']==winner_name].loser_elo.iloc[i:]
    sub_wl = df[df['winner_name']==loser_name].winner_elo.iloc[i:]
    
    sub_ww.replace(to_replace=sub_ww.values.tolist(), value=elo_winner, inplace=True)
    sub_ll.replace(to_replace=sub_ll.values.tolist(), value=elo_loser, inplace=True)
    sub_lw.replace(to_replace=sub_lw.values.tolist(), value=elo_winner, inplace=True)
    sub_wl.replace(to_replace=sub_wl.values.tolist(), value=elo_loser, inplace=True)
    
    df.update(sub_ww)
    df.update(sub_ll)
    df.update(sub_lw)
    df.update(sub_wl)
```
